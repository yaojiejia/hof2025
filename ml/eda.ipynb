```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
plt.scatter(df['score'], df['change'], s=20, color='orange', edgecolors='blue', linewidth=0.5)
plt.xlabel('Number of comments')
plt.ylabel('Change in stock price')
plt.title('Number of upvotes on the comment vs Change in price')
plt.show()
```

```python
plt.figure()
plt.scatter(df['sentiment_score'], df['change'], s=20, color='orange', edgecolors='blue', linewidth=0.5)
plt.xlabel('Sentiment score')
plt.ylabel('Change in stock price')
plt.title('Sentiment Score vs Change in stock prices')
plt.show()
```
